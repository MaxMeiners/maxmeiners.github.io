```python
import pandas as pd

df = pd.read_excel("green_index.xlsx", decimal=",")
df
```

```python
import folium
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd

# Create a GeoDataFrame from the DataFrame with Point geometry
geometry = [Point(xy) for xy in zip(df['longitude'], df['latitude'])]
gdf = gpd.GeoDataFrame(df, geometry=geometry)

# Set the CRS for the GeoDataFrame
gdf.crs = 'EPSG:4326'

# Create a Folium map centered around the average latitude and longitude
map_center = [gdf['latitude'].mean(), gdf['longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=10)

# Create a GeoJSON layer from the GeoDataFrame
folium.GeoJson(
    gdf,
    name='GeoJSON',
    popup=folium.GeoJsonPopup(
        fields=['year', 'green_score', 'date'],
        aliases=['Year', 'Green Score', 'Date'],
        localize=True,
        labels=True,
        style='background-color: yellow;'
    )
).add_to(m)

# Display the map
m
```

```python
import folium
from branca.colormap import linear

df_2015 = df[df["year"] == 2015]

# Get minimum and maximum values of green_score
min_score = df_2015['green_score'].min()
max_score = df_2015['green_score'].max()

# Create a linear color scale
colormap = linear.RdYlGn_11.scale(min_score, max_score)

map = folium.Map(location=[51.5831, 4.7762], zoom_start=12)

# Loop through the rows of the filtered DataFrame and add a marker to the map for each row
for index, row in df_2015.iterrows():
    folium.Marker([row['latitude'], row['longitude']], 
                  # Use green_score to set marker color
                  icon=folium.Icon(color=colormap(row['green_score'])), 
                  popup=row['green_score']).add_to(map)
    
# Add color scale to the map
colormap.caption = 'Green Score'
map.add_child(colormap)

map.save('map_2015.html')
```
