```python
import pandas as pd
```

```python
df_population = pd.read_csv("./datasets/population/population_breda_neighborhoods.csv", sep=";")
df_population = df_population.rename(columns={"Unnamed: 0": "Neighbourhood"})
```

```python
df_population
```

```python
df_population_T = df_population.rename(columns={"Neighbourhood": "Year"})
df_population_T = df_population_T.transpose()
new_header = df_population_T.iloc[0]
df_population_T = df_population_T[1:]
df_population_T.columns = new_header
df_population_T
```

```python
df_nuisance = pd.read_csv("./datasets/registered_nuisance/registered_nuisance.csv", sep=";")

nuisance_key_title = pd.read_csv("./datasets/registered_nuisance/nuisance_key_title.csv", sep=";")
neighbourhood_key_title = pd.read_csv("./datasets/registered_nuisance/neighbourhood_key_title.csv", sep=";")

df_nuisance = pd.merge(df_nuisance, nuisance_key_title, left_on="Overlast", right_on="Key")
df_nuisance= df_nuisance.rename(columns={"Title": "Nuisance"})
df_nuisance = df_nuisance.drop("Key", axis=1)

df_nuisance = pd.merge(df_nuisance, neighbourhood_key_title, left_on="WijkenEnBuurten", right_on="Key")
df_nuisance= df_nuisance.rename(columns={"Title": "Neighbourhood", "GeregistreerdeOverlast_1": "Reports_number"})
df_nuisance = df_nuisance.drop("Key", axis=1)

df_nuisance = df_nuisance.drop(columns=["Overlast", "WijkenEnBuurten"])
```

```python
df_nuisance_total_by_year = df_nuisance[(df_nuisance["Nuisance"] == "Totaal registraties overlast") & (df_nuisance["Perioden"].str[-2:] == "00")]
df_nuisance_total_by_year["Year"] = df_nuisance_total_by_year["Perioden"].str[:4]
df_nuisance_total_by_year
```

```python
df_population_by_year = df_population.melt(id_vars=["Neighbourhood"], var_name="Year", value_name="Population")

# Use next year's population
df_population_by_year["Year"] = (df_population_by_year['Year'].astype(int) + 1).astype(str)
```

```python
df_merged = df_population_by_year.merge(df_nuisance_total_by_year, on=["Neighbourhood", "Year"])

df_merged
```

# Baseline ML model

### Based on the current population of a neighbourhood, can we predict the number of total registered nuisance reports in the next year?

```python
df_train = df_merged[df_merged["Year"] < "2020"]
df_test = df_merged[df_merged["Year"] >= "2020"]

X_train = df_train[["Neighbourhood", "Population"]]
X_test = df_test[["Neighbourhood", "Population"]]
X_train = pd.get_dummies(X_train, columns=["Neighbourhood"], prefix=["N"])
X_test = pd.get_dummies(X_test, columns=["Neighbourhood"], prefix=["N"])
y_train = df_train["Reports_number"]
y_test = df_test["Reports_number"]

y_train = y_train.values.ravel()
y_test = y_test.values.ravel()
```

```python
# from sklearn.preprocessing import StandardScaler

# scaler = StandardScaler()
# X['Reports_number'] = scaler.fit_transform(X['Reports_number'].values.reshape(-1, 1))
```

```python
from sklearn.linear_model import LinearRegression

# Initialize the linear regression model
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions on the test data
y_pred = model.predict(X_test)

# Evaluate model performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-Squared Score:", r2)
```
