```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

# Features Extraction


### Population


```python
df_population = pd.read_csv(
    "./datasets/population/population_breda_neighborhoods.csv", sep=";"
)
df_population = df_population.rename(columns={"Unnamed: 0": "Neighbourhood"})
df_population_by_year = df_population.melt(
    id_vars=["Neighbourhood"], var_name="Year", value_name="Population"
)
```

### Unemployment


```python
uitkering_2022 = pd.read_csv(
    "./datasets/personen_met_uitkering_wijk_buurt/2022_breda.csv", sep=";"
)
uitkering_2021 = pd.read_csv(
    "./datasets/personen_met_uitkering_wijk_buurt/2021_breda.csv", sep=";"
)
uitkering_2020 = pd.read_csv(
    "./datasets/personen_met_uitkering_wijk_buurt/2020_breda.csv", sep=";"
)
uitkering_2019 = pd.read_csv(
    "./datasets/personen_met_uitkering_wijk_buurt/2019_breda.csv", sep=";"
)
uitkering_2018 = pd.read_csv(
    "./datasets/personen_met_uitkering_wijk_buurt/2018_breda.csv", sep=";"
)
uitkering_2017 = pd.read_csv(
    "./datasets/personen_met_uitkering_wijk_buurt/2017_breda.csv", sep=";"
)
uitkering_2016 = pd.read_csv(
    "./datasets/personen_met_uitkering_wijk_buurt/2016_breda.csv", sep=";"
)
uitkering_2015 = pd.read_csv(
    "./datasets/personen_met_uitkering_wijk_buurt/2015_breda.csv", sep=";"
)
uitkering_2014 = pd.read_csv(
    "./datasets/personen_met_uitkering_wijk_buurt/2014_breda.csv", sep=";"
)
uitkering_2013 = pd.read_csv(
    "./datasets/personen_met_uitkering_wijk_buurt/2013_breda.csv", sep=";"
)
```

```python
len(uitkering_2022.columns), len(uitkering_2021.columns), len(
    uitkering_2020.columns
), len(uitkering_2019.columns), len(uitkering_2018.columns), len(
    uitkering_2017.columns
), len(
    uitkering_2016.columns
), len(
    uitkering_2015.columns
), len(
    uitkering_2014.columns
), len(
    uitkering_2013.columns
)
```

```python
df_uitkering = pd.concat(
    [
        uitkering_2022,
        uitkering_2021,
        uitkering_2020,
        uitkering_2019,
        uitkering_2018,
        uitkering_2017,
        uitkering_2016,
        uitkering_2015,
        uitkering_2014,
        uitkering_2013,
    ]
)
```

```python
df_uitkering["Year"] = df_uitkering["Perioden"].str[:4]
df_uitkering["Year"].unique()
```

```python
df_uitkering
```

```python
df_uitkering[
    "Personen per soort uitkering/Werkloosheidsuitkering (aantal)"
] = df_uitkering[
    "Personen per soort uitkering/Werkloosheidsuitkering (aantal)"
].replace(
    "       .", np.nan
)
```

```python
df_uitkering = df_uitkering.dropna(
    subset=["Personen per soort uitkering/Werkloosheidsuitkering (aantal)"]
)
```

```python
df_uitkering[
    "Personen per soort uitkering/Werkloosheidsuitkering (aantal)"
] = df_uitkering["Personen per soort uitkering/Werkloosheidsuitkering (aantal)"].astype(
    int
)
```

```python
df_uitkering = (
    df_uitkering.groupby(["Wijken en buurten", "Year"])[
        "Personen per soort uitkering/Werkloosheidsuitkering (aantal)"
    ]
    .mean()
    .reset_index()
)
```

```python
df_uitkering = df_uitkering.rename(
    columns={
        "Wijken en buurten": "Neighbourhood",
        "Personen per soort uitkering/Werkloosheidsuitkering (aantal)": "Unemployment",
    }
)
```

```python
df_uitkering["Unemployment"] = df_uitkering["Unemployment"].astype(int)
```

```python
df_uitkering
```

### Nuisance


```python
df_nuisance = pd.read_csv(
    "./datasets/registered_nuisance/registered_nuisance.csv", sep=";"
)

nuisance_key_title = pd.read_csv(
    "./datasets/registered_nuisance/nuisance_key_title.csv", sep=";"
)
neighbourhood_key_title = pd.read_csv(
    "./datasets/registered_nuisance/neighbourhood_key_title.csv", sep=";"
)

df_nuisance = pd.merge(
    df_nuisance, nuisance_key_title, left_on="Overlast", right_on="Key"
)
df_nuisance = df_nuisance.rename(columns={"Title": "Nuisance"})
df_nuisance = df_nuisance.drop("Key", axis=1)

df_nuisance = pd.merge(
    df_nuisance, neighbourhood_key_title, left_on="WijkenEnBuurten", right_on="Key"
)
df_nuisance = df_nuisance.rename(
    columns={"Title": "Neighbourhood", "GeregistreerdeOverlast_1": "Reports_number"}
)
df_nuisance = df_nuisance.drop("Key", axis=1)

df_nuisance = df_nuisance.drop(columns=["Overlast", "WijkenEnBuurten"])
```

```python
df_nuisance
```

```python
df_nuisance_total_by_year = df_nuisance[
    (df_nuisance["Nuisance"] == "Totaal registraties overlast")
    & (df_nuisance["Perioden"].str[-2:] == "00")
]
df_nuisance_total_by_year["Year"] = df_nuisance_total_by_year["Perioden"].str[:4]
df_nuisance_total_by_year
```

```python
df_nuisance_total_by_year = df_nuisance_total_by_year[
    ["Neighbourhood", "Year", "Reports_number"]
]
```

```python
df_nuisance_total_by_year
```

```python
df_nextYearNuisance = df_nuisance_total_by_year.copy()
df_nextYearNuisance["Year"] = df_nextYearNuisance["Year"].astype(int) - 1
df_nextYearNuisance["Year"] = df_nextYearNuisance["Year"].astype(str)
```

```python
df_nextYearNuisance = df_nextYearNuisance.rename(
    columns={"Reports_number": "Total Nuisance Next Year"}
)
```

```python
df_nextYearNuisance
```

### Crime


```python
df_crime = pd.read_csv("./datasets/registered_crime/registered_crime.csv", sep=";")
```

```python
df_crime = df_crime[df_crime["Soort misdrijf"] == "Totaal misdrijven"]
```

```python
df_crime["Year"] = df_crime["Perioden"].str[:4]
```

```python
df_crime
```

```python
# df_crime = df_crime.dropna(subset=["Geregistreerde misdrijven (aantal)"])
```

```python
df_crime["Geregistreerde misdrijven (aantal)"] = df_crime[
    "Geregistreerde misdrijven (aantal)"
].astype(int)
```

```python
df_crime_by_year = (
    df_crime.groupby(["Wijken en buurten", "Year"])[
        "Geregistreerde misdrijven (aantal)"
    ]
    .sum()
    .reset_index()
)
```

```python
df_crime_by_year
```

# Merging


```python
df_crime_by_year = df_crime_by_year.rename(
    columns={"Wijken en buurten": "Neighbourhood"}
)
```

```python
df_crime_nuisance = pd.merge(
    df_crime_by_year, df_nextYearNuisance, on=["Year", "Neighbourhood"]
)
```

```python
df_crime_nuisance = df_crime_nuisance.rename(
    columns={
        "Geregistreerde misdrijven (aantal)": "Total crimes",
    }
)
```

```python
df_crime_nuisance
```

```python
df_crime_nuisance_unemployment = pd.merge(
    df_crime_nuisance, df_uitkering, on=["Year", "Neighbourhood"]
)
```

```python
df_crime_nuisance_unemployment
```

```python
df_merged = pd.merge(
    df_crime_nuisance_unemployment, df_population_by_year, on=["Year", "Neighbourhood"]
)
```
