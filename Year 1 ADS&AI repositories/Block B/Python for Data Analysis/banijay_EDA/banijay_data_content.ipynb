## Content data preparation

### Check and remove missing data

```python
df_content.isnull().values.any().sum()
```

### Check and remove duplicate data (duplicated across all columns)

```python
df_content.drop_duplicates
```

### Check the format of the start and end columns and convert them to datetime format.

```python
df_content.info() #Checking the format of the data.
```

```python
df_content['start'] = df_content['start'].map(lambda x: str(x)[:-3]) #Removing the milliseconds from the start column.
```

```python
df_content.tail(2) #Checking if the milliseconds were removed.
```

```python
df_content['end'] = df_content['end'].map(lambda x: str(x)[:-3]) #Removing the milliseconds from the end column.
```

```python
df_content.tail(2) #Checking if the milliseconds were removed.
```

### Create a date_time_start and date_time_end column by combining the date and time from the Datum, start and end columns.

```python
df_content['date_time_start'] = pd.to_datetime(df_content['date'] + ' ' + df_content['start'], infer_datetime_format=True)
df_content['date_time_end'] = pd.to_datetime(df_content['date'] + ' ' + df_content['end'], infer_datetime_format=True)
df_content.head(2)
```

### Split the id column into show_id and fragment columns where fragments can be 1,2,3,or 4.

```python
df_content[['show_id','fragment']] = df_content.id.str.split("_0", expand=True)
df_content.head(2) #Checking if splitting the id column worked.
```

```python
df_content.info() #Checking the data types of the columns to see if date_time_start and date_time_end are datetime64[ns].
```

### Use the to_csv() function to save the dataframe as a CSV file and name it banijay_op1data_content_processed.csv and upload/commit the changes to your Github repos.

```python
df_content.to_csv('data/banijay_op1data_content_processed.csv', compression='zip') #Compressed the data to be able to upload it to GitHub.
```
